```python
def expand_contractions(text, contraction_mapping=contractions_dict):
    
    contractions_pattern = re.compile('({})'.format('|'.join(contraction_mapping.keys())), 
                                      flags=re.IGNORECASE|re.DOTALL)
    def expand_match(contraction):
        match = contraction.group(0)
        first_char = match[0]
        expanded_contraction = contraction_mapping.get(match)\
                                if contraction_mapping.get(match)\
                                else contraction_mapping.get(match.lower())  
        while not AttributeError:
            try:
                expanded_contraction = first_char+expanded_contraction[1:]
            except:
                continue
            return expanded_contraction
        
    expanded_text = contractions_pattern.sub(expand_match, text)
    expanded_text = re.sub("'", "", expanded_text)
    return expanded_text
```

#### XG Boost

```python
metric_xgb = pd.DataFrame()
future_pred_window=1
for l in range(3,15):
    #Using seg-learn to create time window for prediction
    segmenter = SegmentXYForecast(l, step=1, y_func=last,forecast=future_pred_window)
    X_train_rolled, y_train_rolled,_=segmenter.fit_transform([X_train_scaled],[y_train_scaled.flatten()])
    X_test_rolled, y_test_rolled,_=segmenter.fit_transform([X_test_scaled],[y_test_scaled.flatten()])
        #Flattening the train
    shape = X_train_rolled.shape
    X_train_flattened = X_train_rolled.reshape(shape[0],shape[1]*shape[2])
    X_train_flattened.shape
        #Flattening the test
    shape = X_test_rolled.shape
    X_test_flattened = X_test_rolled.reshape(shape[0],shape[1]*shape[2])
    X_test_flattened.shape
    n_est=np.arange(100,1000,50)
    mse_n_est=[]
    for i in n_est:
              model_tuned=XGBRegressor(random_state=1000,
                             n_estimators=i).fit(X_train_flattened,y_train_rolled)
              pred=model_tuned.predict(X_test_flattened)
              mse_n_est.append(metrics.mean_squared_error(y_test_rolled,pred))
    learning_r=[0.01,0.05,0.1,0.15,0.2,0.3]
    mse_lr=[]
    for i in learning_r:
              model_tuned=XGBRegressor(random_state=1000,
              learning_rate=i,).fit(X_train_flattened,y_train_rolled)
              pred=model_tuned.predict(X_test_flattened)
              mse_lr.append(metrics.mean_squared_error(y_test_rolled,pred))
        #Optimised Model
    model_xgb_tuned=XGBRegressor(learning_rate=learning_r[mse_lr.index(min(mse_lr))],random_state=4000,n_estimators=n_est[mse_n_est.index(min(mse_n_est))],
                    max_depth=2).fit(X_train_flattened,y_train_rolled)
    pred_train=model_xgb_tuned.predict(X_train_flattened)
    pred_test=model_xgb_tuned.predict(X_test_flattened)
    mse_train=metrics.mean_squared_error(y_train[l:],pred_train)
    mse_test=metrics.mean_squared_error(y_test[l:],pred_test)
    mae_train=metrics.mean_squared_error(y_train[l:],pred_train)
    mae_test=metrics.mean_squared_error(y_test[l:],pred_test)
    metric_xgb_tuned={'Classical Model':['XGboost Regressor_Tuned'],
                      'Time window' : l,
                      'MSE Train':mse_train,
                      'MSE Test':mse_test,
                      'MAE Train':mae_train,
                      'MAE Test':mae_test}
    metric_xgb_tune=pd.DataFrame(metric_xgb_tuned)
    metric_xgb = pd.concat([metric_xgb_tune,metric_xgb],ignore_index=True)
    metric_xgb
```

```python
metric_lstm_f = pd.DataFrame()
future_pred_window=1
for l in range(3,15):
    segmenter = SegmentXYForecast(l, step=1, y_func=last,forecast=future_pred_window)
    X_train_rolled, y_train_rolled,_=segmenter.fit_transform([X_train_scaled],[y_train_scaled.flatten()])
    X_test_rolled, y_test_rolled,_=segmenter.fit_transform([X_test_scaled],[y_test_scaled.flatten()])
        #Flattening the train
    shape = X_train_rolled.shape
    X_train_flattened = X_train_rolled.reshape(shape[0],shape[1]*shape[2])
    X_train_flattened.shape
        #Flattening the test
    shape = X_test_rolled.shape
    X_test_flattened = X_test_rolled.reshape(shape[0],shape[1]*shape[2])
    X_test_flattened.shape
    # Building the LSTM model here!
    model_lstm = Sequential()
    model_lstm.add(LSTM(units=LSTM_CELL_SIZE,activation = 'relu',input_shape=(l,4)))
    model_lstm.add(Dense (5, activation = 'relu'))
    model_lstm.add(Dropout(DROPOUT_RATE))
    model_lstm.add(Dense (1))
    model_lstm.compile(loss='mean_squared_error', optimizer=Adam(learning_rate = 0.02),metrics=["accuracy"])
    history = model_lstm.fit(X_train_rolled, y_train_rolled,validation_data=(X_test_rolled, y_test_rolled), epochs=EPOCHS,batch_size=BATCH_SIZE,verbose =1)
    pred_train=model_lstm.predict(X_train_rolled)
    pred_test=model_lstm.predict(X_test_rolled)
    mse_train=metrics.mean_squared_error(y_train[l:],pred_train)
    mse_test=metrics.mean_squared_error(y_test[l:],pred_test)
    mae_train=metrics.mean_squared_error(y_train[l:],pred_train)
    mae_test=metrics.mean_squared_error(y_test[l:],pred_test)
    metric_lstm={'Model':['LSTM'],
            "Time Window" : l,
            'MSE Train':mse_train,
            'MAE Train':mae_train,
            'MSE Test':mse_test,
            'MAE Test':mae_test}
    metric_lstm=pd.DataFrame(metric_lstm)
    metric_lstm_f = pd.concat([metric_lstm,metric_lstm_f],ignore_index=True)
    metric_lstm_f
```

```python
temp=df_merged.groupby(['Date'])
pos_avg=[]
neg_avg=[]
neu_avg=[]
close=[]
std_Threshold = 0.1

for date in df_merged.Date.unique():
    Temp = temp.get_group(date)
    close.append(Temp['ClosePrice'].iat[0])
    if Temp['NumOfTweets'].iat[0] == 1:
        pos_avg.append(Temp['Positive'].iat[0])
        neg_avg.append(Temp['Negative'].iat[0])
        neu_avg.append(Temp['Neutral'].iat[0])
        
    else:
        if ((Temp['Std_Positive'].iat[0] < std_Threshold) & (Temp['Std_Negative'].iat[0] < std_Threshold) & (Temp['Std_Neutral'].iat[0] < std_Threshold)):
            pos_avg.append(Temp['Positive'].mean())
            neg_avg.append(Temp['Negative'].mean())
            neu_avg.append(Temp['Neutral'].mean())
        else:
            pos_avg.append(Temp['Positive'][Temp['Tweet_Engagement']==Temp['Tweet_Engagement'].max()].iat[0])
            neg_avg.append(Temp['Negative'][Temp['Tweet_Engagement']==Temp['Tweet_Engagement'].max()].iat[0])
            neu_avg.append(Temp['Neutral'][Temp['Tweet_Engagement']==Temp['Tweet_Engagement'].max()].iat[0])
```

```python
metric_lstm_sh = pd.DataFrame()
future_pred_window=1
for l in range(3,15):
    segmenter = SegmentXYForecast(l, step=1, y_func=last,forecast=future_pred_window)
    X_train_rolled, y_train_rolled,_=segmenter.fit_transform([X_train_scaled],[y_train_scaled.flatten()])
    X_test_rolled, y_test_rolled,_=segmenter.fit_transform([X_test_scaled],[y_test_scaled.flatten()])
        #Flattening the train
    shape = X_train_rolled.shape
    X_train_flattened = X_train_rolled.reshape(shape[0],shape[1]*shape[2])
    X_train_flattened.shape
        #Flattening the test
    shape = X_test_rolled.shape
    X_test_flattened = X_test_rolled.reshape(shape[0],shape[1]*shape[2])
    X_test_flattened.shape
    # Building the LSTM model here!
    model_lstm = Sequential()
    model_lstm.add(LSTM(units=LSTM_CELL_SIZE,activation = 'relu',input_shape=(l,4)))
    model_lstm.add(Dense (5, activation = 'relu'))
    model_lstm.add(Dropout(DROPOUT_RATE))
    model_lstm.add(Dense (1))
    model_lstm.compile(loss='mean_squared_error', optimizer=Adam(learning_rate = 0.02),metrics=["accuracy"])
    history = model_lstm.fit(X_train_rolled, y_train_rolled,validation_data=(X_test_rolled, y_test_rolled), epochs=EPOCHS,batch_size=BATCH_SIZE,verbose =1)
    pred_train=model_lstm.predict(X_train_rolled)
    pred_test=model_lstm.predict(X_test_rolled)
    mse_train=metrics.mean_squared_error(y_train[l:],pred_train)
    mse_test=metrics.mean_squared_error(y_test[l:],pred_test)
    mae_train=metrics.mean_squared_error(y_train[l:],pred_train)
    mae_test=metrics.mean_squared_error(y_test[l:],pred_test)
    metric_lstm={'Model':['LSTM'],
            "Time Window" : l,
            'MSE Train':mse_train,
            'MAE Train':mae_train,
            'MSE Test':mse_test,
            'MAE Test':mae_test}
    metric_lstm=pd.DataFrame(metric_lstm)
    metric_lstm_sh = pd.concat([metric_lstm,metric_lstm_sh],ignore_index=True)
    metric_lstm_sh
```
